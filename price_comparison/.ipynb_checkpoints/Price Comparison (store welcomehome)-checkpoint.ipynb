```python
import pandas as pd
import numpy as np
```

#### Read Data

```python
shopee = pd.read_csv('Input/Shopee.csv')
shopee
```

```python
lazada = pd.read_csv('Input/Lazada.csv')
lazada
```

#### Prerpocess Data

```python
name_shopee = shopee["Productname"]
name_lazada = lazada["name"]

from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import EnglishStemmer
from nltk import word_tokenize
import re
import string

def stem_tokens(t,s):
    stemmed = []
    for item in t:
        stemmed.append(s.stem(item))
    return stemmed

def tokenize(text):
    #stemmer = PorterStemmer()
    stemmer = EnglishStemmer()
    text = re.sub(r"[^a-zA-Z0-9]"," ",text)
    
    tokens = word_tokenize(text)
    tokens = [i for i in tokens if i not in string.punctuation]
    
    stems = stem_tokens(tokens, stemmer)
    return stems

count_vectorizer = CountVectorizer(tokenizer = tokenize, lowercase = True)
shopee_product = count_vectorizer.fit_transform(name_shopee)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer(use_idf=True)
shopee_product_tfidf = tfidf_transformer.fit_transform(shopee_product)
```

```python
lazada_product = count_vectorizer.transform(name_lazada)
lazada_product_tfidf = tfidf_transformer.transform(lazada_product)
```

#### Output Result

```python
columns = ["shopee_product_ID","shopee_name", "shopee_cat","shopee_price","lazada_name","lazada_cat","lazada_price","lazada_discount_price","lazada_weight","similarity_score"]
output_dataframe = pd.DataFrame(columns=columns)
output_dataframe
```

```python
lazada_array = lazada_product_tfidf.toarray()
for i in range(0,len(name_lazada)):
    result = shopee_product_tfidf.dot(lazada_array[i])
    shopee_index = np.argmax(result)
    shopee_row = shopee.iloc[shopee_index,:]
    lazada_row = lazada.iloc[i,:]
    output_dataframe = output_dataframe.append(pd.DataFrame([[shopee_row["ProductID"], shopee_row["Productname"],shopee_row["Category"],shopee_row["Price"],lazada_row["name"],lazada_row["category_level1"],lazada_row["price_discount"],lazada_row["price_discount"],lazada_row["weight"],result[shopee_index]]],columns = columns),ignore_index=True)
```

```python
output_dataframe.to_csv('welcomehome_product_match.csv')
```
