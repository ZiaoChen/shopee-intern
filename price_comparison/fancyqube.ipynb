#### Create Term Document Frequency Matrix

```python
name_shopee = shopee["Productname"]
name_lazada = lazada["name"].apply(lambda x: x.split("\n",1)[0])

from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import EnglishStemmer
from nltk import word_tokenize
from nltk.util import ngrams
import re
import string

def stem_tokens(t,s):
    stemmed = []
    for item in t:
        stemmed.append(s.stem(item))
    return stemmed

def tokenize(text):
    #stemmer = PorterStemmer()
    stemmer = EnglishStemmer()
    text = re.sub(r"[^a-zA-Z0-9]"," ",text)
    
    tokens = word_tokenize(text)
    tokens = [i for i in tokens if i not in string.punctuation]
    
    stems = stem_tokens(tokens, stemmer)
        
    return stems

count_vectorizer = CountVectorizer(tokenizer = tokenize, lowercase = True)
lazada_product = count_vectorizer.fit_transform(name_lazada)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer(use_idf=True)
lazada_product_tfidf = tfidf_transformer.fit_transform(lazada_product)
```

```python
shopee_product = count_vectorizer.transform(name_shopee)
shopee_product_tfidf = tfidf_transformer.transform(shopee_product)
```
